<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:20px;">6.Key Insights

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:18px;"> > Attendance acts as the "Floor." There is a clear "Success Threshold" at 75%. Below this point, exam scores drop sharply regardless of how many hours a student studies at home.Classroom engagement provides the structure needed for study hours to be effective.

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:18px;"> > Internet access is a Performance Multiplier. Students without internet access face a "Statistical Ceiling," rarely breaking into the top 10% of scorers ($>85\%$), even with high motivation levels.Lack of digital resources limits a student's ability to research, use educational software, or access remote help.

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:18px;"> > Parental involvement serves as an "Academic Safety Net." High involvement doesn't always lead to an "A," but it significantly reduces the "Variance" in scores, effectively preventing students from bottoming out into the "Fail" category.Support systems provide accountability and emotional stability, keeping students in the "Safe Zone" even during low-motivation periods.

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:18px;"> > Failure is rarely caused by a single factor. It is usually the result of a "Negative Chain" (e.g., No Internet + Low Attendance + Low Motivation). If a student has all three risks, the probability of failure is nearly 80%.Positive variables (like high motivation) can sometimes overcome one negative factor, but they cannot overcome three.

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:20px;">7.Conclusion

<p style="color:MidnightBlue;font-family:ConicSansMS;font-size:18px;">The analysis of these 6,600+ records demonstrates that student performance is built on a "success-first" foundation, where a staggering 99% pass rate makes success the baseline and failure a rare anomaly. While internal factors like Motivation and Attendance are the strongest predictors of top-tier scores, environmental supports—specifically Internet Access and Parental Involvement—act as essential stabilizers that prevent at-risk students from falling behind. Moving forward, the project's value lies not in predicting success, which is almost guaranteed, but in isolating the specific "perfect storm" of negative variables that impact the 1% minority of failing students.
